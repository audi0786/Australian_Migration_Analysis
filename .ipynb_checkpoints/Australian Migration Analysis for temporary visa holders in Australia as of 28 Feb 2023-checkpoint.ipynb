These datasets contain information about various temporary visa subclasses and their holders in Australia.

The first dataset explores the trends of temporary visa holders across different Australian states, while the second dataset provides detailed information on various temporary visa subclasses and their respective numbers at different points in time.

Through our analysis, we aim to identify patterns and insights in the Australian temporary visa landscape, and gain a better understanding of the trends and anomalies in the data. This information can be valuable for policymakers, researchers, and anyone interested in the dynamics of temporary migration in Australia.

```python
import pandas as pd

# Replace 'your_data.csv' with your actual data file
data = pd.read_csv('457_Decomposition.csv', encoding='latin1')

# Set the column names
column_names = [
    'Nominated Position Location (State)',
    'Nominated Occupation (Major Group)',
    '3/31/2013', '6/30/2013', '9/30/2013', '12/31/2013',
    '3/31/2014', '6/30/2014', '9/30/2014', '12/31/2014',
    '3/31/2015', '6/30/2015', '9/30/2015', '12/31/2015',
    '3/31/2016', '6/30/2016', '9/30/2016', '12/31/2016',
    '3/31/2017', '6/30/2017', '9/30/2017', '12/31/2017',
    '3/31/2018', '6/30/2018', '9/30/2018', '12/31/2018',
    '3/31/2019', '6/30/2019', '9/30/2019', '12/31/2019',
    '3/31/2020', '6/30/2020', '9/30/2020', '12/31/2020',
    '3/31/2021', '6/30/2021', '9/30/2021', '12/31/2021',
    '3/31/2022', '6/30/2022', '9/30/2022', '12/31/2022'
]
data.columns = column_names

# Display the first 5 rows of the dataset
data.head()
```

```python
data_long = pd.melt(
    data,
    id_vars=['Nominated Position Location (State)', 'Nominated Occupation (Major Group)'],
    var_name='Snapshot Date',
    value_name='Total'
)

data_long.head()
```

```python
data_long['Nominated Occupation (Major Group)'] = data_long['Nominated Occupation (Major Group)'].replace('Not Applicable', 'Secondary Applicants')
# Not applicable stands for secondary applicants who are not supposed to have any nominated occupation for their visa.
```

```python
data_long['Snapshot Date'] = pd.to_datetime(data_long['Snapshot Date'])
```

```python
data_long.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

# Create a list of occupation groups
occupation_groups = data_long['Nominated Occupation (Major Group)'].dropna().unique()

# Determine the number of rows and columns for the grid
n_rows = int(np.ceil(len(occupation_groups) / 3))

# Set up the grid
fig, axes = plt.subplots(n_rows, 3, figsize=(25, 6 * n_rows), sharex=False, sharey=False)
axes = axes.flatten()

# Loop through each occupation group and create a time series plot for each one
for i, occupation_group in enumerate(occupation_groups):
    data_subset = data_long[data_long['Nominated Occupation (Major Group)'] == occupation_group]
    sns.lineplot(data=data_subset, x='Snapshot Date', y='Total', ax=axes[i], marker='o')
    axes[i].set_title(occupation_group)
    axes[i].set_ylabel('Total Applicants', fontsize=14)
    axes[i].set_xlabel('Snapshot Date', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45, labelsize=10)
    axes[i].tick_params(axis='y', rotation=45, labelsize=12)
    axes[i].yaxis.set_major_locator(ticker.MaxNLocator(6)) # Set the maximum number of y-axis ticks

# Remove unused subplots
for i in range(len(occupation_groups), n_rows * 3):
    fig.delaxes(axes[i])
    

# Adjust spacing between plots
plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Add a title for the entire grid
fig.suptitle('Total Applicants Over Time by Occupation (Major Group)', fontsize=25, y=0.9)

plt.show()


```

```python
# Filter data for Queensland
data_long_queensland = data_long[data_long['Nominated Position Location (State)'] == 'QLD']

# Convert the 'Total' column to a numeric data type
data_long_queensland['Total'] = pd.to_numeric(data_long_queensland['Total'], errors='coerce')

# Create a time series plot
plt.figure(figsize=(12, 8))
sns.lineplot(data=data_long_queensland, x='Snapshot Date', y='Total', hue='Nominated Occupation (Major Group)', marker='o')
plt.title('Total Applicants Over Time by Occupation Major Group in Queensland', fontsize=24)
plt.ylabel('Total Applicants', fontsize=16)
plt.xlabel('Snapshot Date', fontsize=16)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1), title='Nominated Occupation (Major Group)', fontsize=9, title_fontsize=11)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(np.arange(0, data_long_queensland['Total'].max() + 500, 500), fontsize=12)
plt.ylim(0, data_long_queensland['Total'].max() + 500)
plt.show()
```

## Australian Temporary Visa Holders Analysis

There are several categories of temporary visas in Australia, including Temporary Activity (408), Temporary Work (Skilled) (457), Work and Holiday (462), Temporary Skill Shortage (482), Temporary Graduate (485), Student (500), Visitor (600), Electronic Travel Authority (601), and Bridging Visas.

1. **Student Visa (500)** consistently holds the highest number of visa holders across the analyzed period. The total number of Student Visa holders increased from 516,574 in June 2019 to 585,785 in February 2023. This growth suggests a strong demand for Australian education among international students.

2. **Visitor Visa (600)** and **Electronic Travel Authority (601)** categories experienced significant fluctuations during the analyzed period. A sharp decline in visa holders can be observed between December 2019 and March 2021, likely due to travel restrictions imposed during the COVID-19 pandemic. However, these categories have shown a steady recovery since then, with the number of Visitor Visa holders increasing from 27,407 in March 2021 to 209,427 in February 2023, and Electronic Travel Authority holders growing from 590 to 66,943 during the same period.

3. **Temporary Graduate Visa (485)** category showed consistent growth over the period, increasing from 91,776 in June 2019 to 167,404 in February 2023. This trend suggests that more international students are opting to stay and work in Australia after completing their studies.

4. **Temporary Work (Skilled) (457)** category experienced a significant decline in visa holders, dropping from 92,158 in June 2019 to 833 in February 2023. This decrease is likely due to the introduction of the Temporary Skill Shortage (482) visa, which has grown steadily, with the number of visa holders increasing from 50,670 in June 2019 to 122,094 in February 2023.

5. **Work and Holiday (462)** category displayed an initial decline in visa holders until March 2021, followed by a strong recovery, with the number of visa holders increasing from 4,609 in March 2021 to 21,603 in February 2023. This resurgence could be attributed to the easing of travel restrictions and a renewed interest in working holidays in Australia.

6. The number of **Bridging Visa** holders increased from 179,921 in June 2019 to 183,200 in February 2023, indicating a growing backlog in visa processing or delays in transitioning to other visa categories.

In summary, the Australian temporary visa landscape has experienced some significant changes during the analyzed period. The growth in Student and Temporary Graduate visas suggests that Australia remains an attractive destination for international students, while the recovery of Visitor and Work and Holiday visas indicates a resurgence in tourism and working holidays.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

grants_refusals = pd.read_csv("Grants_Refusals_PointsBasedVisas.csv")
on_hand_visas = pd.read_csv("On_hand_visas_pointsBased.csv")
lodged_visas = pd.read_csv("Number_of_lodged_visas.csv")
```

```python
granted = grants_refusals[grants_refusals['Outcome'] == 'Granted'].iloc[:, 2:].fillna(0)
refused = grants_refusals[grants_refusals['Outcome'] == 'Refused'].iloc[:, 2:].fillna(0)
visa_subclasses = grants_refusals['Visa Subclass'].unique()

fig, axes = plt.subplots(len(visa_subclasses), 1, figsize=(12, 20), sharex=True)

for i, subclass in enumerate(visa_subclasses):
    ax = axes[i]
    ind = np.arange(len(granted.columns))
    width = 0.35

    grant_row = granted.loc[grants_refusals['Visa Subclass'] == subclass]
    refuse_row = refused.loc[grants_refusals['Visa Subclass'] == subclass]
    
    ax.bar(ind - width/2, grant_row.values[0] if len(grant_row) > 0 else [0] * len(grant_row.columns), width, label='Granted')
    ax.bar(ind + width/2, refuse_row.values[0] if len(refuse_row) > 0 else [0] * len(refuse_row.columns), width, label='Refused')
    ax.set_title(f'{subclass} - Granted vs Refused Visas')
    ax.set_xticks(ind)
    ax.set_xticklabels(granted.columns)
    ax.legend()

plt.show()

```

```python
fig, ax = plt.subplots(figsize=(12, 6))
on_hand_visas.set_index("Visa Subclass").T.plot(ax=ax)
ax.set_title("On-hand Visas per Subclass")
ax.set_xlabel("Months")
ax.set_ylabel("Number of Visas")
ax.legend(title="Visa Subclass", loc="upper left")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
lodged_visas.set_index("Visa Subclass").T.plot(ax=ax)
ax.set_title("Number of Lodged Visas per Subclass")
ax.set_xlabel("Months")
ax.set_ylabel("Number of Visas")
ax.legend(title="Visa Subclass", loc="upper left")
plt.show()
```
