```python
import pandas as pd

# Replace 'your_data.csv' with your actual data file
data = pd.read_csv('457_Decomposition.csv', encoding='latin1')

# Set the column names
column_names = [
    'Nominated Position Location (State)',
    'Nominated Occupation (Major Group)',
    '3/31/2013', '6/30/2013', '9/30/2013', '12/31/2013',
    '3/31/2014', '6/30/2014', '9/30/2014', '12/31/2014',
    '3/31/2015', '6/30/2015', '9/30/2015', '12/31/2015',
    '3/31/2016', '6/30/2016', '9/30/2016', '12/31/2016',
    '3/31/2017', '6/30/2017', '9/30/2017', '12/31/2017',
    '3/31/2018', '6/30/2018', '9/30/2018', '12/31/2018',
    '3/31/2019', '6/30/2019', '9/30/2019', '12/31/2019',
    '3/31/2020', '6/30/2020', '9/30/2020', '12/31/2020',
    '3/31/2021', '6/30/2021', '9/30/2021', '12/31/2021',
    '3/31/2022', '6/30/2022', '9/30/2022', '12/31/2022'
]
data.columns = column_names

# Display the first 5 rows of the dataset
data.head()
```

```python
data_long = pd.melt(
    data,
    id_vars=['Nominated Position Location (State)', 'Nominated Occupation (Major Group)'],
    var_name='Snapshot Date',
    value_name='Total'
)

data_long.head()
```

```python
data_long['Snapshot Date'] = pd.to_datetime(data_long['Snapshot Date'])
```

```python
data_long.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.lineplot(data=data_long, x='Snapshot Date', y='Total', hue='Nominated Occupation (Major Group)')
plt.title('Total Applicants Over Time by Occupation Major Group')
plt.ylabel('Total Applicants')
plt.show()
```

```python
import numpy as np
def convert_to_float(x):
    try:
        return float(str(x).replace(',', ''))
    except ValueError:
        return np.nan

data_long['Total'] = data_long['Total'].apply(convert_to_float)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data_long, x='Nominated Position Location (State)', y='Total', estimator=sum, ci=None)
plt.title('Total Applicants by State')
plt.ylabel('Total Applicants')
plt.xlabel('Nominated Position Location (State)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=data_long, x='Nominated Occupation (Major Group)', y='Total', estimator=sum, ci=None)
plt.title('Total Applicants by Occupation Major Group')
plt.ylabel('Total Applicants')
plt.xlabel('Nominated Occupation (Major Group)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(18, 8))
sns.boxplot(data=data_long, x='Nominated Position Location (State)', y='Total', hue='Nominated Occupation (Major Group)')
plt.title('Applicants Distribution by State and Occupation Major Group')
plt.ylabel('Total Applicants')
plt.xlabel('Nominated Position Location (State)')
plt.xticks(rotation=45)
plt.show()
```

```python
heatmap_data = data_long.pivot_table(
    index='Nominated Position Location (State)',
    columns='Nominated Occupation (Major Group)',
    values='Total',
    aggfunc='mean'
)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, fmt='.2f', cmap='YlGnBu', cbar_kws={'label': 'Average Applicants'})
plt.title('Average Applicants per Snapshot Date by State and Occupation Major Group')
plt.ylabel('Nominated Position Location (State)')
plt.xlabel('Nominated Occupation (Major Group)')
plt.show()
```

## Australian Temporary Visa Holders Analysis

There are several categories of temporary visas in Australia, including Temporary Activity (408), Temporary Work (Skilled) (457), Work and Holiday (462), Temporary Skill Shortage (482), Temporary Graduate (485), Student (500), Visitor (600), Electronic Travel Authority (601), and Bridging Visas.

1. **Student Visa (500)** consistently holds the highest number of visa holders across the analyzed period. The total number of Student Visa holders increased from 516,574 in June 2019 to 585,785 in February 2023. This growth suggests a strong demand for Australian education among international students.

2. **Visitor Visa (600)** and **Electronic Travel Authority (601)** categories experienced significant fluctuations during the analyzed period. A sharp decline in visa holders can be observed between December 2019 and March 2021, likely due to travel restrictions imposed during the COVID-19 pandemic. However, these categories have shown a steady recovery since then, with the number of Visitor Visa holders increasing from 27,407 in March 2021 to 209,427 in February 2023, and Electronic Travel Authority holders growing from 590 to 66,943 during the same period.

3. **Temporary Graduate Visa (485)** category showed consistent growth over the period, increasing from 91,776 in June 2019 to 167,404 in February 2023. This trend suggests that more international students are opting to stay and work in Australia after completing their studies.

4. **Temporary Work (Skilled) (457)** category experienced a significant decline in visa holders, dropping from 92,158 in June 2019 to 833 in February 2023. This decrease is likely due to the introduction of the Temporary Skill Shortage (482) visa, which has grown steadily, with the number of visa holders increasing from 50,670 in June 2019 to 122,094 in February 2023.

5. **Work and Holiday (462)** category displayed an initial decline in visa holders until March 2021, followed by a strong recovery, with the number of visa holders increasing from 4,609 in March 2021 to 21,603 in February 2023. This resurgence could be attributed to the easing of travel restrictions and a renewed interest in working holidays in Australia.

6. The number of **Bridging Visa** holders increased from 179,921 in June 2019 to 183,200 in February 2023, indicating a growing backlog in visa processing or delays in transitioning to other visa categories.

In summary, the Australian temporary visa landscape has experienced some significant changes during the analyzed period. The growth in Student and Temporary Graduate visas suggests that Australia remains an attractive destination for international students, while the recovery of Visitor and Work and Holiday visas indicates a resurgence in tourism and working holidays.
